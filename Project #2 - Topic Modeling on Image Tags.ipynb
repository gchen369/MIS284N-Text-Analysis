```python
import pandas as pd
from sklearn import preprocessing

min_max_scaler = preprocessing.MinMaxScaler()

df['likes_scaled']= min_max_scaler.fit_transform(df[['likes']])
df['comments_scaled']= min_max_scaler.fit_transform(df[['comments']])
df = df.drop(df.columns[0], axis=1)
df['engagement'] = 0.4*df['likes_scaled'] + 0.6*df['comments_scaled']
df.head()
```

```python
import gensim
from gensim.utils import simple_preprocess
from gensim.parsing.preprocessing import STOPWORDS
from nltk.stem import WordNetLemmatizer, SnowballStemmer
from nltk.stem.porter import *
import numpy as np
import nltk
nltk.download('wordnet')
```

```python
stemmer = PorterStemmer()
def lemmatize_stemming(text):
    return stemmer.stem(WordNetLemmatizer().lemmatize(text, pos='v'))
def preprocess(text):
    text= ' '.join(text)
    result = []
    for token in gensim.utils.simple_preprocess(text):
        if token not in gensim.parsing.preprocessing.STOPWORDS and len(token) > 0:
            result.append(lemmatize_stemming(token))
    return result
```

```python
doc_sample = df['labels'].iloc[0]
doc_sample = ' '.join(doc_sample)

print('original document: ')
words = []
for word in doc_sample.split(' '):
    words.append(word)
print(words)
print('\n\n tokenized and lemmatized document: ')
print(preprocess(doc_sample))
```

```python
processed_docs = df['labels'].map(preprocess)
processed_docs[:10]
```

```python
dictionary = gensim.corpora.Dictionary(processed_docs)
count = 0
for k, v in dictionary.iteritems():
    print(k, v)
    count += 1
    if count > 10:
        break
```

```python
dictionary.filter_extremes(no_below=15, no_above=0.5, keep_n=100000)
```

```python
bow_corpus = [dictionary.doc2bow(doc) for doc in processed_docs]
bow_corpus[5]
```

```python
bow_doc_5 = bow_corpus[5]
for i in range(len(bow_doc_5)):
    print("Word {} (\"{}\") appears {} time.".format(bow_doc_5[i][0], 
                                               dictionary[bow_doc_5[i][0]], 
bow_doc_5[i][1]))
```

```python
from gensim import corpora, models
tfidf = models.TfidfModel(bow_corpus)
corpus_tfidf = tfidf[bow_corpus]
from pprint import pprint
for doc in corpus_tfidf[:5]:
    pprint(doc)
```

```python
lda_model = gensim.models.LdaMulticore(bow_corpus, num_topics = 5, id2word=dictionary, passes=2, workers=2)
for idx, topic in lda_model.print_topics(-1):
    print('Topic: {} \nWords: {}'.format(idx, topic))
```

```python
lda_model_tfidf = gensim.models.LdaMulticore(corpus_tfidf, num_topics=5, id2word=dictionary, passes=2, workers=4)
for idx, topic in lda_model_tfidf.print_topics(-1):
    print('Topic: {} Word: {}'.format(idx, topic))
```

```python
unseen_document = df['labels'].iloc[19]
bow_vector = dictionary.doc2bow(preprocess(unseen_document))
for index, score in sorted(lda_model_tfidf[bow_vector], key=lambda tup: -1*tup[1]):
    print("Score: {}\t Topic: {}".format(score, lda_model_tfidf.print_topic(index, 5)))
```

```python
def topic_weights(x):
    bow_vector = dictionary.doc2bow(preprocess(unseen_document))
    return lda_model_tfidf[bow_vector]
```

```python
df['topic_weights'] = df['labels'].copy()
df['topic_weights'] = df['topic_weights'].map(topic_weights)
```

```python
high_quart = df[df['engagement'] > np.quantile(df['engagement'], .75)]
low_quart = df[df['engagement'] < np.quantile(df['engagement'], .25)]
```

```python
counth0,counth1,counth2,counth3,counth4 = 0,0,0,0,0
for x in high_quart['topic_weights']:
    for y in x:
        if y[0] == 0:
            #top0+=1
            #print(y)
            #print(y[0])
            counth0+=y[1]
        elif y[0] == 1:
            #top1+=1
            counth1+=y[1]
        elif y[0] == 2:
            #top2+=1
            counth2+=y[1]
        elif y[0] == 3:
            #top3+=1
            counth3+=y[1]
        elif y[0] == 4:
            #top4+=1
            counth4+=y[1]
print('Average Engagment on the high quartile for Topic 1: ', counth0/121)
print('Average Engagment on the high quartile for Topic 2: ', counth1/121)
print('Average Engagment on the high quartile for Topic 3: ', counth2/121)
print('Average Engagment on the high quartile for Topic 4: ', counth3/121)
print('Average Engagment on the high quartile for Topic 5: ', counth4/121)
```

```python
count0,count1,count2,count3,count4 = 0,0,0,0,0
top0 = 0
for x in low_quart['topic_weights']:
    for y in x:
        if y[0] == 0:
            top0+=1
            count0+=y[1]
        elif y[0] == 1:
            #top1+=1
            count1+=y[1]
        elif y[0] == 2:
            #top2+=1
            count2+=y[1]
        elif y[0] == 3:
            #top3+=1
            count3+=y[1]
        elif y[0] == 4:
            #top1+=1
            count4+=y[1]
print(top0)
print('Average Engagment on the low quartile for Topic 1: ', count0/121)
print('Average Engagment on the low quartile for Topic 2: ', count1/121)
print('Average Engagment on the low quartile for Topic 3: ', count2/121)
print('Average Engagment on the low quartile for Topic 4: ', count3/121)
print('Average Engagment on the low quartile for Topic 5: ', count4/121)
```
